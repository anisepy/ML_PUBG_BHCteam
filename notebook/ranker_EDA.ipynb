### DataFrame 생성(kaggle crawling overlap)

```python
def calculate_quantile(dataframe, columns_name, kaggle=df):
    x = kaggle.copy()
    describe = dataframe.describe()
    for i in x.columns:
        if i in columns_name.columns:
            # print(i)
            # df = df[df["assists"] > ranker_describe.loc["25%"]["assists"]][df["assists"] < ranker_describe.loc["75%"]["assists"]]
            x = x[x[i] >= describe.loc["25%"][i]][x[i] <= describe.loc["75%"][i]]
    IQR_1 = x.winPlacePerc.quantile(.25)
    IQR_3 = x.winPlacePerc.quantile(.75)
    return describe,IQR_1,IQR_3,x
```

### 각각 feature별 winPlacePerc에 영향끼치는지 구하기(IQR)

```python
def each_calculate_quantile(dataframe):
    x = df.copy()
    describe = dataframe.describe()
    total = []
    for i in x.columns:
        if i in dataframe.columns:
#             print(i)
            each_column = []
            column_name = "ranker_" + i
            each_column.append(column_name)
            column_name = x[x[i] >= describe.loc["25%"][i]][x[i] <= describe.loc["75%"][i]]
            each_column.append(column_name.winPlacePerc.quantile(.25).item())
            each_column.append(column_name.winPlacePerc.quantile(.75).item())
            total.append(each_column)
    total_df = pd.DataFrame(total, columns=["columns_name","Min","Max"])
    return total_df
```

```python
def boxplot(columns_name, df, overlap_df):
    plt.figure(figsize=(10,12))
    plt.subplot(2,1,1)
    a = str(columns_name)
    sns.boxplot(overlap_df[columns_name])
    plt.xlim([0,df[columns_name].max()])
    plt.subplot(2,1,2)
    sns.boxplot(df[columns_name])
    plt.xlim([0,df[columns_name].max()])
    plt.show()
#     sns.boxplot()
```

```python
def countplot(columns_name, df, overlap_df):
    plt.figure(figsize=(10,12))
    plt.subplot(2,1,1)
    a = str(columns_name)
    sns.countplot(overlap_df[columns_name])
    plt.xlim([0.5,df[columns_name].max()+1])
    plt.subplot(2,1,2)
    sns.countplot(df[columns_name])
    plt.xlim([0.5,df[columns_name].max()+1])
    plt.show()
```

```python
def barplot(columns_name, df, overlap_df):
    plt.figure(figsize=(10,12))
    plt.subplot(2,1,1)
    a = str(columns_name)
    sns.barplot(overlap_df[columns_name])
    plt.xlim([-1,df[columns_name].max()])
    plt.subplot(2,1,2)
    sns.barplot(df[columns_name])
    plt.xlim([-1,df[columns_name].max()])
    plt.show()
```

```python
def distplot(columns_name,df, overlap_df,hist=True):
    plt.figure(figsize=(10,12))
    plt.subplot(2,1,1)
    a = str(columns_name)
    sns.distplot(overlap_df[columns_name],hist=hist)
    plt.xlim([-1,df[columns_name].max()])
    plt.subplot(2,1,2)
    sns.distplot(df[columns_name], hist=hist)
    plt.xlim([-1,df[columns_name].max()])
    plt.show()
```

```python
def scatterplot(columns_name,alpha=0.03):
    plt.figure(figsize=(10,12))
    plt.subplot(2,1,1)
    a = str(columns_name)
    sns.scatterplot(overlap_df[columns_name],overlap_df.winPlacePerc,alpha=alpha)
    plt.xlim([-1,df[columns_name].max()])
    plt.ylim([0,1])
    plt.subplot(2,1,2)
    sns.scatterplot(df[columns_name],df.winPlacePerc,alpha=alpha)
    plt.xlim([-1,df[columns_name].max()])
    plt.ylim([0,1])
    plt.show()
```
